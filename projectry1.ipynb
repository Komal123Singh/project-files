```python
#Python-Project
import MySQLdb
from tkinter import *
import tkinter.messagebox as messagebox
from tkinter import ttk
import datetime
import time

class AA:
    def addcustomer(self):
        self.fname=self.first.get()
        self.flast=self.last.get()
        self.femail=self.email.get()
        self.fmobile=self.mobile.get()
        db=MySQLdb.connect("localhost","root","root","alphaprojectdb")
        cursor=db.cursor()
        sqlquery="insert into customer(firstname,lastname,email,mobileno) values(%s,%s,%s,%s)"
        print("S")
        if(self.fname =="" or self.flast =="" or self.femail== "" or self.fmobile==""):
            messagebox.showerror("Update Status","All Fields are mandatory")
        elif(self.fmobile.isdigit()=='False' or len(self.fmobile)!=10):
            messagebox.showerror("Wrong", "Mobile No. must be numeric and length of 10")
        else:
            try:
                cursor.execute(sqlquery,(self.fname,self.flast,self.femail,self.fmobile))
                data=cursor.fetchall()
                print(data)
                print("Successfully ")
                db.commit()
            except:
                db.rollback()
            db.close()
            print("Successfully added")
            messagebox.showinfo("Success","customer added")

    def add(self):
        win1=Tk()
        win1.title('Add Customer')
        win1.geometry('430x365+500+200')
        win1.configure(background="white")
        
        frame=LabelFrame(win1,text="Add New Record")
        frame.grid(row=0,column=0)
        
        Label(frame,text='Firstname').grid(row=1,column=1)
        self.first=Entry(frame,bd=4)
        self.first.grid(row=1,column=2)
        
        Label(frame,text='Lastname').grid(row=2,column=1)
        self.last=Entry(frame,bd=4)
        self.last.grid(row=2,column=2)
        
        Label(frame,text='Email').grid(row=3,column=1)
        self.email=Entry(frame,bd=4)
        self.email.grid(row=3,column=2)
        
        Label(frame,text='Mobile no.').grid(row=4,column=1)
        self.mobile=Entry(frame,bd=4)
        self.mobile.grid(row=4,column=2)
        
        ttk.Button(frame,text='Add Record',command=self.addcustomer).grid(row=5,column=2)
        self.message=Label(text='',fg='red')
        self.message.grid(row=6,column=1)
        
    def delete(self):
        win2=Tk()
        win2.title('Delete Customer')
        win2.geometry('430x365+500+200')
    
        frame=LabelFrame(win2,text="Search Record")
        frame.grid(row=0,column=0)
        
        
        ttk.Button(frame,text='Delete Record').grid(row=4,column=2)
        self.message=Label(text='',fg='red')
        self.message.grid(row=5,column=1)
        
    def __init__(self,win):
        self.lbl1=Label(win,text="Welcome to Alpha Inventory System...!",font=('arial',15,'bold'))
        self.lbl1.grid(row=0,column=1)

#-------Menu Bar--------------------------------
        chooser=Menu()
       
        itemthree=Menu(tearoff=0)
        itemthree.add_command(label='Add Customer',command=self.add)
        itemthree.add_command(label='Delete Customer',command=self.delete)
        
        chooser.add_cascade(label='Customer Master',menu=itemthree)
        win.config(menu=chooser)
        
if __name__ == '__main__':
    window = Tk()
    alphaobj = AA(window)
    window.title('Alpha Inventory System...!')
    window.geometry('1350x768')
    window.mainloop()
```
