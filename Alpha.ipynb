```python
from tkinter import*
from tkinter import ttk
import random
from tkinter import messagebox
from tkinter import filedialog as fd
import MySQLdb

class AlphaSystem:
    def __init__(self,win):
        self.win=win

        self.win.title("Alpha Inventory System...!!!")
        self.win.geometry("500x465+500+200")
        self.win.configure(background='powder blue')
         
             
        chooser=Menu()
    
        itemone=Menu(tearoff=0)
        itemone.add_command(label='Add Product' ,command =self.addproduct)
        itemone.add_command(label='View Product')
        itemone.add_command(label='Delete Product')
        itemone.add_separator()
        itemone.add_command(label='Help')
        
        chooser.add_cascade(label='Product Master',menu=itemone)
        win.config(menu=chooser)
    
    def browseImage(self):
        filename = fd.askopenfilename()
        print((filename))
        self.image.insert(END,filename)
        
    
    def convertBinaryFun(self,filename):
        with open(filename,"rb") as f:
            binaryData = f.read()
        return binaryData
        
    def insertData(self):
        self.pid=int(self.id.get())
        self.pname=self.name.get()
        self.pcategory=self.category.get()
        self.pquantity=int(self.quantity.get())
        self.pprice=float(self.price.get())
        self.pdescription=self.description.get()
        self.pimage=self.image.get()
        
        if self.pid==0:
            messagebox.showerror("Error","Please enter product name.....?")
        elif self.pname=='':
            messagebox.showerror("Error","Please enter name first.....?")
        elif self.pcategory=='':
            messagebox.showerror("Error","please enter category first.....?")
        elif self.pquantity==0:
            messagebox.showerror("Error","Please enter quantity first.....?")
        elif self.pprice==0:
            messagebox.showerror("Error","please enter price first.....?")    
        elif self.pdescription=='':
            messagebox.showerror("Error","Please enter desciption first.....?")
        elif self.pimage=='':
            messagebox.showerror("Error","please enter image first.....?") 
            
        else:   
            try:
                db = MySQLdb.connect('localhost','root','root123','alphaprojectdb')

                #cursor object creation
                cursor = db.cursor()
                sqlquery = "insert into productmaster(Pid,Name,Category,Quantity,Price,Description,Image) values(%s,%s,%s,%s,%s,%s,%s)"
                binimage = self.convertBinaryFun(self.image.get())


                #executing sql query
                res=cursor.execute(sqlquery,(self.pid,self.pname,self.pcategory,self.pquantity,self.pprice,self.pdescription,binimage))
                #commit the changes into a database table
                db.commit()
                print("Record inserted successfully.....!")
                messagebox.showinfo("Success","Record has been saved succesfully.")
            except MySQLdb.Error as error:
                print('Failed to insert the record with image',error)
                db.rollback()
            db.close()
    
    
    def addproduct(self):
        
        self.addproduct = Toplevel(win)
        self.addproduct.title("ADD PRODUCT") 
        self.addproduct.geometry("500x465+500+200")
        self.title=Label(self.addproduct,text='ADD PRODUCT',font=("times new roman",20,"bold"),bg='cyan')
        self.title.place(x=0,y=0,relwidth=1)
        
        
        self.lblid=Label(self.addproduct,text='Pid',font=('arial',12))
        self.lblid.place(x=20,y=50)
        self.id=Entry(self.addproduct)
        self.id.place(x=110,y=50)
        
        self.lblname=Label(self.addproduct,text='Name',font=('arial',12))
        self.lblname.place(x=20,y=100)
        self.name=Entry(self.addproduct)
        self.name.place(x=110,y=100)
        
        
        self.lblcategory=Label(self.addproduct,text='Category',font=('arial',12))
        self.lblcategory.place(x=20,y=150)
        self.category=Entry(self.addproduct)
        self.category.place(x=110,y=150)
        
        
        self.lblquantity=Label(self.addproduct,text='Quantity',font=('arial',12))
        self.lblquantity.place(x=20,y=200)
        self.quantity=Entry(self.addproduct)
        self.quantity.place(x=110,y=200)
        
        self.lblprice=Label(self.addproduct,text='Price',font=('arial',12))
        self.lblprice.place(x=20,y=250)
        self.price=Entry(self.addproduct)
        self.price.place(x=110,y=250)
        

        self.lbldescription=Label(self.addproduct,text='Description',font=('arial',12))
        self.lbldescription.place(x=20,y=300)
        self.description=Entry(self.addproduct)
        self.description.place(x=110,y=300)
        
        
        self.lblimage=Label(self.addproduct,text='Image',font=('arial',12))
        self.lblimage.place(x=20,y=350)
        self.image=Entry(self.addproduct)
        self.image.place(x=110,y=350)
        self.imagebtn=Button(self.addproduct,text='Select Image',command=self.browseImage)
        self.imagebtn.place(x=250,y=350)
        
        self.submitbtn=Button(self.addproduct,text='SUBMIT',command=self.insertData)
        self.submitbtn.place(x=220,y=400)


if __name__ == '__main__':
    win = Tk()
    obj=AlphaSystem(win)
    win.mainloop()   
```
